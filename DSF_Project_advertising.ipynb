```python
df = pd.read_csv('/content/advertising.csv')
```

```python
print(df.shape)
print(df.columns)
print(df.sample(6))
```

```python
print(df.describe())
```

```python
fig = pyplot.figure (figsize =(8,6))
ax = Axes3D(fig)

x = df['TV']
y = df['Radio']
z = df['Sales']

ax.scatter(x, y, z, c= 'r')
ax.set_xlabel('Presupuesto TV')
ax.set_ylabel('Presupuesto Radio')
ax.set_zlabel('Ventas')
```

Ajustar modelo de regresin mediante formulas

```python
modelo = smf.ols('z ~ x + y', data=df).fit()
print(modelo.summary())
```
